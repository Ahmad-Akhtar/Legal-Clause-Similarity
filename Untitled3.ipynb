```python
# Import Libraries & Configration
```

```python
import os
import glob
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras import layers, models, backend as K
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import nltk
from nltk.corpus import stopwords
import re
import json

# reproducibility
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# nltk downloads (run once)
nltk.download('punkt')
nltk.download('stopwords')

# -------------------------------------------------------------------------------
# 2. Configuration / hyperparameters


CONFIG = {
    "max_vocab_size": 30000,
    "max_seq_len": 100,
    "embedding_dim": 128,
    "lstm_units": 128,
    "batch_size": 16,
    "epochs": 12,
    "validation_split": 0.1,
    "test_size": 0.15,
    "oov_token": "<OOV>"
}
```

```python
# Import Data set
```

```python
def load_dataset_from_folder(folder_path):
    import glob, os
    import pandas as pd

    files = glob.glob(os.path.join(folder_path, "*.csv"))
    if not files:
        raise ValueError(f"No CSV files found in {folder_path}")

    dfs = []
    for f in files:
        df = pd.read_csv(f)
        # Check expected columns
        if 'clause_text' not in df.columns or 'clause_type' not in df.columns:
            raise ValueError(f"File {f} does not have required columns ['clause_text', 'clause_type']")
        df = df[['clause_text', 'clause_type']].rename(columns={'clause_text': 'text', 'clause_type': 'label'})
        df['source_file'] = os.path.basename(f)
        dfs.append(df)

    big = pd.concat(dfs, ignore_index=True)
    big['text'] = big['text'].astype(str)
    big['label'] = big['label'].astype(str)
    return big
```

```python
# Text cleaning and preprocessing
```

```python
STOPWORDS = set(stopwords.words('english'))

def clean_text(s):
    s = s.lower()
    s = re.sub(r'\s+', ' ', s)
    s = re.sub(r'http\S+|www.\S+', '', s)          # remove urls
    s = re.sub(r'[^a-z0-9\s\.,;:\'\"()-]', ' ', s) # remove weird chars but keep punctuation
    s = s.strip()
    return s

def preprocess_df(df):
    df = df.copy()
    df['text_clean'] = df['text'].apply(clean_text)
    # Optionally remove stopwords or do more sophisticated preprocessing
    return df
```

```python
# Pair generation (Siamese-style)
```

```python
def make_pairs(df, max_pos_pairs_per_label=1000, max_neg_pairs=None):
    """
    Create positive and negative pairs:
    - positive: two clauses with same label
    - negative: two clauses with different labels
    Returns DataFrame with columns: text_a, text_b, label (1=similar,0=different)
    """
    label_groups = df.groupby('label').apply(lambda x: x.index.tolist()).to_dict()
    indices = df.index.tolist()
    pos_pairs = []
    for label, idxs in label_groups.items():
        if len(idxs) < 2:
            continue
        # create combinations but limit count per label
        pairs = []
        for i in range(len(idxs)):
            for j in range(i+1, len(idxs)):
                pairs.append((idxs[i], idxs[j]))
        random.shuffle(pairs)
        pairs = pairs[:max_pos_pairs_per_label]
        for i, j in pairs:
            pos_pairs.append((df.at[i,'text_clean'], df.at[j,'text_clean'], 1))
    # negative pairs - sample randomly across labels
    neg_pairs = []
    num_neg = max_neg_pairs if max_neg_pairs is not None else len(pos_pairs)
    while len(neg_pairs) < num_neg:
        a, b = random.sample(indices, 2)
        if df.at[a,'label'] != df.at[b,'label']:
            neg_pairs.append((df.at[a,'text_clean'], df.at[b,'text_clean'], 0))
    pairs = pos_pairs + neg_pairs
    random.shuffle(pairs)
    pairs_df = pd.DataFrame(pairs, columns=['text_a','text_b','label'])
    return pairs_df
```

```python
# Tokenization and sequence encoding
```

```python
def build_tokenizer(texts, max_vocab=CONFIG['max_vocab_size'], oov_token=CONFIG['oov_token']):
    tok = Tokenizer(num_words=max_vocab, oov_token=oov_token)
    tok.fit_on_texts(texts)
    return tok

def texts_to_padded_sequences(tokenizer, texts, maxlen=CONFIG['max_seq_len']):
    seqs = tokenizer.texts_to_sequences(texts)
    padded = pad_sequences(seqs, maxlen=maxlen, padding='post', truncating='post')
    return padded
```

```python
# Model architectures
```

```python
def attention_layer(inputs):
    """Simple attention mechanism over time dimension.
    inputs: shape (batch, timesteps, features)
    returns: context vector (batch, features)
    """
    # compute attention scores
    score_dense = layers.Dense(1, activation='tanh')
    scores = score_dense(inputs)  # (batch, timesteps, 1)
    scores = layers.Flatten()(scores)  # (batch, timesteps)
    alphas = layers.Activation('softmax')(scores)  # (batch, timesteps)
    alphas = layers.RepeatVector(inputs.shape[-1])(alphas)  # (features, timesteps) permutation next
    alphas = layers.Permute([2,1])(alphas)  # (batch, timesteps, features)
    weighted = layers.multiply([inputs, alphas])
    context = layers.Lambda(lambda x: K.sum(x, axis=1))(weighted)
    return context

def build_shared_encoder(vocab_size, embedding_dim=CONFIG['embedding_dim'], max_seq_len=CONFIG['max_seq_len'],
                         lstm_units=CONFIG['lstm_units'], use_attention=False, name='shared_encoder'):
    inp = layers.Input(shape=(max_seq_len,), name=f'{name}_input')
    x = layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_seq_len, mask_zero=True)(inp)
    x = layers.Bidirectional(layers.LSTM(lstm_units, return_sequences=True))(x)
    if use_attention:
        x = attention_layer(x)
    else:
        # use global max/average pooling
        x1 = layers.GlobalMaxPooling1D()(x)
        x2 = layers.GlobalAveragePooling1D()(x)
        x = layers.Concatenate()([x1, x2])  # combined representation
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    encoder_model = models.Model(inputs=inp, outputs=x, name=name)
    return encoder_model

def build_siamese_model(vocab_size, use_attention=False):
    shared = build_shared_encoder(vocab_size, use_attention=use_attention)
    a_in = layers.Input(shape=(CONFIG['max_seq_len'],), name='A_input')
    b_in = layers.Input(shape=(CONFIG['max_seq_len'],), name='B_input')
    a_vec = shared(a_in)
    b_vec = shared(b_in)
    # combine
    diff = layers.Lambda(lambda tensors: K.abs(tensors[0] - tensors[1]))([a_vec, b_vec])
    mult = layers.multiply([a_vec, b_vec])
    merged = layers.Concatenate()([diff, mult, a_vec, b_vec])
    x = layers.Dense(256, activation='relu')(merged)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(64, activation='relu')(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    model = models.Model(inputs=[a_in, b_in], outputs=out)
    return model
```

```python
# Training & evaluation helpers
```

```python
def compile_and_train(model, X_train, y_train, X_val, y_val, model_name='model'):
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    ckpt_path = f"{model_name}.h5"
    callbacks = [
        EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
        ModelCheckpoint(ckpt_path, save_best_only=True, monitor='val_loss')
    ]
    history = model.fit(
        [X_train['a'], X_train['b']], y_train,
        validation_data=([X_val['a'], X_val['b']], y_val),
        batch_size=CONFIG['batch_size'],
        epochs=CONFIG['epochs'],
        callbacks=callbacks,
        verbose=2
    )
    return history, ckpt_path

def evaluate_model(model, X_test, y_test):
    y_pred_prob = model.predict([X_test['a'], X_test['b']]).ravel()
    y_pred = (y_pred_prob >= 0.5).astype(int)
    acc = accuracy_score(y_test, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='binary', zero_division=0)
    roc_auc = None
    try:
        roc_auc = roc_auc_score(y_test, y_pred_prob)
    except Exception:
        roc_auc = float('nan')
    metrics = {'accuracy': acc, 'precision': precision, 'recall': recall, 'f1': f1, 'roc_auc': roc_auc}
    return metrics, y_pred, y_pred_prob
```

```python
# Visualization utilities
```

```python
def plot_history(hist, title='Training history'):
    plt.figure(figsize=(12,5))
    # loss
    plt.subplot(1,2,1)
    plt.plot(hist.history['loss'], label='train_loss')
    plt.plot(hist.history['val_loss'], label='val_loss')
    plt.title('Loss')
    plt.legend()
    # accuracy
    plt.subplot(1,2,2)
    plt.plot(hist.history.get('accuracy', []), label='train_acc')
    plt.plot(hist.history.get('val_accuracy', []), label='val_acc')
    plt.title('Accuracy')
    plt.legend()
    plt.suptitle(title)
    plt.show()

def plot_roc(y_test, y_score):
    fpr, tpr, thresholds = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(6,6))
    plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.3f})')
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.show()
```

```python
# Main pipeline
```

```python
def run_pipeline(data_folder):
    # 10.1 Load data
    print("Loading CSV files from", data_folder)
    df = load_dataset_from_folder(data_folder)
    print(f"Loaded {len(df)} clauses from {df['source_file'].nunique()} files, labels: {df['label'].nunique()}")

    # 10.2 Preprocess
    df = preprocess_df(df)
    print("Sample cleaned text:", df['text_clean'].iloc[0])

    # 10.3 Create pairs
    pairs = make_pairs(df, max_pos_pairs_per_label=500, max_neg_pairs=None)
    print("Total pairs:", len(pairs), "Positives:", pairs['label'].sum(), "Negatives:", len(pairs)-pairs['label'].sum())

    # 10.4 Train/validation/test split
    train_df, test_df = train_test_split(pairs, test_size=CONFIG['test_size'], random_state=SEED, stratify=pairs['label'])
    train_df, val_df = train_test_split(train_df, test_size=CONFIG['validation_split'], random_state=SEED, stratify=train_df['label'])
    print("Splits -> train:", len(train_df), "val:", len(val_df), "test:", len(test_df))

    # 10.5 Build tokenizer on all texts
    all_texts = pd.concat([train_df['text_a'], train_df['text_b'], val_df['text_a'], val_df['text_b'], test_df['text_a'], test_df['text_b']])
    tokenizer = build_tokenizer(all_texts.tolist(), max_vocab=CONFIG['max_vocab_size'])
    vocab_size = min(CONFIG['max_vocab_size'], len(tokenizer.word_index)+1)
    print("Vocab size:", vocab_size)

    # 10.6 Encode sequences
    def encode_df(df_):
        a_seq = texts_to_padded_sequences(tokenizer, df_['text_a'].tolist(), CONFIG['max_seq_len'])
        b_seq = texts_to_padded_sequences(tokenizer, df_['text_b'].tolist(), CONFIG['max_seq_len'])
        return {'a': a_seq, 'b': b_seq}

    X_train = encode_df(train_df)
    X_val = encode_df(val_df)
    X_test = encode_df(test_df)
    y_train = train_df['label'].values
    y_val = val_df['label'].values
    y_test = test_df['label'].values

    # 10.7 Build & train Model 1: Siamese BiLSTM (no attention)
    print("\nBuilding Model 1: Siamese BiLSTM (no attention)")
    model1 = build_siamese_model(vocab_size, use_attention=False)
    model1.summary()
    hist1, ckpt1 = compile_and_train(model1, X_train, y_train, X_val, y_val, model_name='siamese_bilstm')
    plot_history(hist1, title='Model 1 Training History')
    metrics1, ypred1, yprob1 = evaluate_model(model1, X_test, y_test)
    print("Model 1 metrics:", metrics1)
    plot_roc(y_test, yprob1)

    # 10.8 Build & train Model 2: Siamese BiLSTM + Attention
    print("\nBuilding Model 2: Siamese BiLSTM + Attention")
    model2 = build_siamese_model(vocab_size, use_attention=True)
    model2.summary()
    hist2, ckpt2 = compile_and_train(model2, X_train, y_train, X_val, y_val, model_name='siamese_bilstm_attention')
    plot_history(hist2, title='Model 2 Training History')
    metrics2, ypred2, yprob2 = evaluate_model(model2, X_test, y_test)
    print("Model 2 metrics:", metrics2)
    plot_roc(y_test, yprob2)

    # 10.9 Comparison summary
    comparison = pd.DataFrame([metrics1, metrics2], index=['Siamese-BiLSTM', 'Siamese-BiLSTM-Attn'])
    display(comparison)

    # 10.10 Show qualitative examples (few correct and incorrect)
    results_df = test_df.copy().reset_index(drop=True)
    results_df['y_true'] = y_test
    results_df['y_pred_prob_model2'] = yprob2
    results_df['y_pred_model2'] = ypred2
    # correct similar predicted
    correct_sim = results_df[(results_df['y_true'] == 1) & (results_df['y_pred_model2'] == 1)].sample(n=min(3, len(results_df[(results_df['y_true'] == 1) & (results_df['y_pred_model2'] == 1)])), random_state=SEED)
    wrong_sim = results_df[(results_df['y_true'] == 1) & (results_df['y_pred_model2'] == 0)].sample(n=min(3, len(results_df[(results_df['y_true'] == 1) & (results_df['y_pred_model2'] == 0)])), random_state=SEED)
    correct_diff = results_df[(results_df['y_true'] == 0) & (results_df['y_pred_model2'] == 0)].sample(n=min(3, len(results_df[(results_df['y_true'] == 0) & (results_df['y_pred_model2'] == 0)])), random_state=SEED)
    wrong_diff = results_df[(results_df['y_true'] == 0) & (results_df['y_pred_model2'] == 1)].sample(n=min(3, len(results_df[(results_df['y_true'] == 0) & (results_df['y_pred_model2'] == 1)])), random_state=SEED)

    def show_examples(df_examples, title):
        print("\n---", title, "---")
        for i, row in df_examples.iterrows():
            print(f"\nA: {row['text_a']}")
            print(f"B: {row['text_b']}")
            print(f"True: {row['y_true']}  Pred_prob: {row['y_pred_prob_model2']:.3f} Pred_label: {row['y_pred_model2']}")

    show_examples(correct_sim, "Correctly predicted similar (model2)")
    show_examples(wrong_sim, "Incorrectly predicted dissimilar but true similar (model2)")
    show_examples(correct_diff, "Correctly predicted different (model2)")
    show_examples(wrong_diff, "Incorrectly predicted similar but true different (model2)")

    # Save tokenizer and config
    with open('tokenizer.json', 'w') as f:
        json.dump(tokenizer.to_json(), f)
    print("\nDone. Models saved as siamese_bilstm.h5 and siamese_bilstm_attention.h5 (if checkpointing succeeded).")
    return {
        'model1': model1,
        'hist1': hist1,
        'model2': model2,
        'hist2': hist2,
        'tokenizer': tokenizer,
        'comparison': comparison
    }

```
